# 1.) Identifying and dropping constant variance features if any

```python
from sklearn.feature_selection import VarianceThreshold
var_thresh = VarianceThreshold(threshold=0.01)   # dropping constant features for now but can increase this threshold to 0.01,0.5,0.1,0.2,0.3,... based on use case
var_thresh.fit(df)

df.columns[var_thresh.get_support()]
```

```python
constant_cols = [col for col in df.columns if col not in df.columns[var_thresh.get_support()]]  #will get list of columns to drop
constant_cols
```

```python
df['Utilities'].value_counts()
```

From above after experimenting with multipe thresholds we can determine 0.01 gives us a reasonable number of columns with low variance which can be seen using value_counts function and hence, they wont be useful for prediction so we drop them

```python
df.drop(constant_cols,axis=1,inplace=True)
df.shape
```

```python
#dropping from test data also
df_test.drop(constant_cols,axis=1,inplace=True)
df_test.shape
```

# 2.) Finding multicollinearity in the input features and dropping such redundant features

```python
fig, ax = plt.subplots(figsize=(40,40))
sns.heatmap(df.corr(),square=True,linewidths=.5,annot=True,vmax=0.3,cmap="coolwarm",cbar=False,xticklabels=df.columns,yticklabels=df.columns) 
```

```python
def multicolin(X, threshold):   # X is the entire train dataste without the output/dependant variable column. Threshold is set as per domain expertise or based on observation.
    col_corr = []
    corr_matrix = X.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                colname1 = corr_matrix.columns[i]
                colname2 = corr_matrix.columns[j]
                col_corr.append((colname1,colname2))

    return col_corr    #will return list of all input feature columns which are highly correlated to each other
```

```python
multicolin_cols = multicolin(df, 0.8)
multicolin_cols
```

```python
from scipy.stats import pearsonr
for col1, col2 in multicolin_cols:
    corr, _ =pearsonr(df[col1],df[col2])
    print(f"Correlation between {col1} and {col2} is {corr}")
```

From above set of correlated input features we will drop one column from each pair to eliminate multicollinearity

```python
cols_to_drop = []
for col1, col2 in multicolin_cols:
    if col1 not in [col2 for col1,col2 in multicolin_cols]:  # this condition added to ensure that a column that maybe present as col1 in tuple x may also be in another pair as col2 in tuple y which may result in complete elimination of tuple y as we eliminate it's col1 regardless and its col2 got removed as part of colx so better to retain a few multicolinear features rather than indadvertantly loosing out on a potentially imp feature
        cols_to_drop.append(col1)
        
set(cols_to_drop)
```

Dropping these columns from train and test data

```python
cols_to_drop = list(set(cols_to_drop))
df.drop(cols_to_drop, axis=1, inplace=True)
df_test.drop(cols_to_drop, axis=1, inplace=True)
```

# 3.) Experimenting with different feature selection techniques and then use reconciled result

## 3.1) Chi Squared Test
Works best with categorical features

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2  #measures dependance between stochaistic variables

#using this we can remove parameters that dont have dependance with target variabble and thus dont provide valuable info for classification.

select_feature=SelectKBest(score_func = chi2, k= 63).fit(df,train['SalePrice'])
dfscore = pd.DataFrame(select_feature.scores_, columns = ['Score']) #scores of all the original features
dfcolumns = pd.DataFrame(df.columns)
features_rank = pd.concat([dfcolumns, dfscore],axis=1)
features_rank.columns = ['Feature','Score']
features_rank     # higher the value of score higher the importance of the feature
```

```python
pd.set_option('display.max_rows',None)
features_rank.sort_values(by=['Score'],ascending=False)
```

## 3.2.) Use ExtraTrees Regressor

```python
from sklearn.ensemble import ExtraTreesRegressor  #dont miss to use Regressor as this is a regression problem
model = ExtraTreesRegressor()
model.fit(df, train['SalePrice'])

print(model.feature_importances_)  # higher the value more important the feature

ranked_features = pd.Series(model.feature_importances_, index=df.columns)
ranked_features
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ranked_features.nlargest(30).plot(kind='barh')
plt.show()  # show top 10 most imp features
```

# 3.3 Mutual Info Gain

```python
from sklearn.feature_selection import mutual_info_regression  #measure dependancy between variables.equal to zero if variables are independant and higher value means higher dependacny. It is non-negative
mutual_info = mutual_info_regression(df,train['SalePrice'])  #use the mutual_info_REGRESSOR
mutual_data = pd.Series(mutual_info, index=df.columns)
mutual_data.sort_values(ascending=False)   #higher value features should be retained and zero value features can be dropped

```

## 5.) RFECV

```python
from sklearn.feature_selection import RFECV
import xgboost

clf_3=xgboost.XGBRegressor()  #using the XgBoost REGRESSOR
rfecv=RFECV(estimator=clf_3,  #estimator being used
           step=1,            #no. of features to be eliminated in each step
           cv=5,              #5-fold CV
           scoring='neg_mean_squared_error', #metric to evaluate model on, since regression problem we use begative mean squared error
           n_jobs=-1).fit(df,train['SalePrice'])         #use all CPU cores
```

```python
print('Optimal number of features: ',rfecv.n_features_)
print('List of optimal features: ',df.columns[rfecv.support_])
```

### Reconciling results of above methods

Approach 1

- From RFECV we can see that only 21 columns are significant. So we take these 21 columns as well as those extra columns from the list of the top 30 columns from mutual_info method and extretrees method which are not already included in the rfecv columns list(to capture any extra variance that may have been missed by the rfecv model) so we get a well distirbuted set of 30-32 columns we can use for our model.

Approach 2

- Take all those columns from mutual info gain model which have a gain value of greater than 0.1

Approach 3

- RFECV showed 21 is optimal number of features so to account for extra variance we will take top 30 columns from ExtraTreesRegressor result

#### Approach 1

```python
rfecv_columns = df.columns[rfecv.support_]
```

```python
mutual_info_columns = mutual_data.sort_values(ascending=False)[:30].index  #taking top 30
```

```python
col_app1 = list(rfecv_columns) + list(extra)
```
